```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
from math import sqrt

col = ["Scn", "A2", "A3", "A4", "A5", "A6", "A7", "A8", "A9", "A10", "Class"]
df = pd.read_csv('Breast-Cancer-Wisconsin.csv', na_values = ['?','NaN'], names = col)
df['A7'].fillna(int(df['A7'].mean()), inplace = True)
df_null_rows = df[df.isnull().any(axis=1)]
df_AColumns = df[["A2", "A3", "A4", "A5", "A6", "A7", "A8", "A9", "A10"]]
df["Predicted_Class"] = 0


# Code for Initial Step
u2_idx = np.random.choice(699, replace = True, size = 1)
u4_idx = np.random.choice(699, replace = True, size = 1)
u2_list = df_AColumns.iloc[u2_idx]
u4_list = df_AColumns.iloc[u4_idx]

print("Randomly selected row " + str(u2_idx[0]) +  " for centroid mu_2.\n")
print("Initial_Centroid mu_2:")
print(u2_list.squeeze())

print("\nRandomly selected row " + str(u4_idx[0]) +  " for centroid mu_4.\n")
print("Initial_Centroid mu_4:")
print(u4_list.squeeze())

u2 = tuple(df_AColumns.iloc[u2_idx].values.flatten().tolist())
u4 = tuple(df_AColumns.iloc[u4_idx].values.flatten().tolist())

iteration = 1 
maxIter = 51
centroidsEqual = False
df_u2_meancentroid =  pd.DataFrame().squeeze()
df_u4_meancentroid =  pd.DataFrame().squeeze()
df_to_compare = df.copy()

while iteration < 51 and not centroidsEqual:
    if(iteration == 1):
        for i in range(len(df_AColumns)):
            # These values are initialized prior to assignment so the value is retained each iteration
            u2_euc = 0 
            u4_euc = 0
            u2_euc_squared = 0
            u4_euc_squared = 0
            if(i >= len(df_AColumns) - 1):
                iteration = iteration + 1 
                df_to_compare = df.copy()
            for j in range(len(df_AColumns.columns)):
                u2_dist_perindex = (df_AColumns.iloc[i][j] - u2[j]) ** 2
                u4_dist_perindex = (df_AColumns.iloc[i][j] - u4[j]) ** 2 
                u2_euc_squared += u2_dist_perindex
                u4_euc_squared += u4_dist_perindex

                if(j>=len(df_AColumns.columns)-1):
                    if(u2_euc_squared < u4_euc_squared):
                        u2_euc = sqrt(u2_euc_squared)
                        u2_rounded = round(sqrt(u2_euc),2)
                        df.at[i,"Predicted_Class"] = 2
                    else:
                        u4_euc =sqrt(u4_euc_squared)              
                        u4_rounded = round(sqrt(u4_euc),2)
                        df.at[i,"Predicted_Class"] = 4

            df_u2_points = df.loc[df['Predicted_Class'] == 2]
            df_u4_points = df.loc[df['Predicted_Class'] == 4]
            df_u2_meancentroid = df_u2_points[["A2", "A3", "A4", "A5", "A6", "A7", "A8", "A9", "A10"]].mean()
            df_u4_meancentroid = df_u4_points[["A2", "A3", "A4", "A5", "A6", "A7", "A8", "A9", "A10"]].mean()

    else:
        #Code for Recompute Step
        for i in range(len(df_AColumns)):
            # These values are initialized prior to assignment so the value is retained each iteration
            u2_euc = 0 
            u4_euc = 0
            u2_euc_squared = 0
            u4_euc_squared = 0
            if(i >= len(df_AColumns) - 1):
                iteration = iteration + 1
                df_u2_mean_previous =  df_u2_meancentroid
                df_u4_mean_previous =  df_u4_meancentroid
                df_u2_points = df.loc[df['Predicted_Class'] == 2]
                df_u4_points = df.loc[df['Predicted_Class'] == 4]
                df_u2_meancentroid = df_u2_points[["A2", "A3", "A4", "A5", "A6", "A7", "A8", "A9", "A10"]].mean()
                df_u4_meancentroid = df_u4_points[["A2", "A3", "A4", "A5", "A6", "A7", "A8", "A9", "A10"]].mean()
                centroidsEqual = df_to_compare.equals(df)
                if(centroidsEqual):
                    break
                else:
                    df_to_compare = df.copy() 
                #should be updated here
          
            for j in range(len(df_AColumns.columns)):
                u2_dist_perindex = (df_AColumns.iloc[i][j] - df_u2_meancentroid[j]) ** 2
                u4_dist_perindex = (df_AColumns.iloc[i][j] - df_u4_meancentroid[j]) ** 2 
                u2_euc_squared += u2_dist_perindex
                u4_euc_squared += u4_dist_perindex

                if(j>=len(df_AColumns.columns)-1):
                    if(u2_euc_squared < u4_euc_squared):
                        u2_euc = sqrt(u2_euc_squared)
                        u2_rounded = round(sqrt(u2_euc),2)
                        df.at[i,"Predicted_Class"] = 2
                    else:
                        u4_euc =sqrt(u4_euc_squared)              
                        u4_rounded = round(sqrt(u4_euc),2)
                        df.at[i,"Predicted_Class"] = 4
        
print("\nProgram ended after " + str(iteration) + " iterations.\n")

print("Final centroid mu_2:")
print(df_u2_meancentroid)

print("\nFinal centroid mu_4:")
print(df_u4_meancentroid)

print("\nFinal cluster assignment:\n")
print(df[["Scn", "Class", "Predicted_Class"]].head(21))




df.to_excel('clusters.xlsx')














```
